```python
def filtred_open_events(df, name_events, name_region):
    name_pattern = '|'.join(name_events)
    df_filtered = df[
        df['Место проведения'].str.contains('Москва', case=False, na=False) &
        df['Наименование мероприятия'].str.contains(name_pattern, case=False, na=False)
    ]
    df_filtered['IASControl'] = 'Не рекомендуем в названии для московских соревнований использовать слова «Открытые» или «Открытый»'
    return df_filtered

filtred_open_events(df, name_events)
```
